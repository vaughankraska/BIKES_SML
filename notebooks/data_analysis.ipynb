```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('../data/training_data.csv')

# Function to plot a feature against bike demand
def plot_feature_vs_demand(data, feature, title):
    plt.figure(figsize=(12, 6))
    sns.countplot(x=feature, hue='increase_stock', data=data)
    plt.title(title)
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Can any trend be seen comparing different hours, weeks, and months?
plot_feature_vs_demand(data, 'hour_of_day', 'Bike Demand by Hour of Day')
plot_feature_vs_demand(data, 'day_of_week', 'Bike Demand by Day of the Week')
plot_feature_vs_demand(data, 'month', 'Bike Demand by Month')
```

```python
# Is there any difference between weekdays and holidays?
plot_feature_vs_demand(data, 'weekday', 'Bike Demand on Weekdays vs Weekends')
plot_feature_vs_demand(data, 'holiday', 'Bike Demand on Holidays vs Regular Days')
```

```python
# Is there any trend depending on the weather? Rainy days, snowy days, etc.
plot_feature_vs_demand(data, 'temp', 'Bike Demand by Temperature')
plot_feature_vs_demand(data, 'precip', 'Bike Demand by Precipitation')
plot_feature_vs_demand(data, 'snow', 'Bike Demand by Snow')
plot_feature_vs_demand(data, 'humidity', 'Bike Demand by Humidity')
plot_feature_vs_demand(data, 'cloudcover', 'Bike Demand by Cloud Cover')
```
